```python
import pandas as pd
import numpy as np
from unidecode import unidecode
```

```python
Dataset=pd.read_csv('jobs (5).csv')
a=Dataset.fillna('NoData') #Cleaning Data
```

```python
withPay=a[(a['Tag 2'].str.contains("Kč")) | ( a['Tag 2'].str.contains("EUR"))]
```

```python
paysparse=withPay['Tag 2'].str.extract(r'(.*0)') #naaah... extracting all the elements that ends up with 0
paysparse=paysparse[paysparse.columns[0]].str.split(" ‍–‍ ", expand = True) #split the bounds
paysparse.columns=['Lower Bound','Upper Bound'] #rename the columns
#paysparse['coruna']=withPay['Tag 2'].str.contains('Kč') #gives True to the line if payment is in Kc
#paykc=paysparse.where(paysparse['coruna']==True).dropna()
paysparse['Lower Bound']=paysparse['Lower Bound'].apply(unidecode).str.replace(' ','').astype(float)
```

```python
import matplotlib.pyplot as plt
%matplotlib widget
fig,ax=plt.subplots()
bins=np.linspace(paysparse['Lower Bound'].min(), paysparse['Lower Bound'].max(), 30)
ax.hist(paysparse['Lower Bound'], density=False, bins=bins)
```
