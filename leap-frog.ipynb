```python
import numpy as np
import pandas as pd
import matplotlib as mpl

from matplotlib import pyplot as plt
%matplotlib ipympl
import scienceplots
```

```python
sol=pd.DataFrame(columns=['x', 'vx','t'])
dt=0.02
T=200
dx=0.05
k=200
m=1
```

$\dot{x}=v_{x_0}$



$v_{i+1/2}=v_{i-1/2}+a \Delta t$

$x_{i+1}=x_{i}+v_{i+1/2}\Delta t$

$v_{i+1/2}=v_{i}+\frac{1}{2}(a_i+a_{i+1}) \Delta t^2$

$x_{i+1}=x_{i}+v_{i}\Delta t+\frac{1}{2}a_i \Delta t$

```python
sol.loc[0]=[10,0,0]
x,vx,t=sol.loc[0]
```

$m\ddot{x}=-kx$

```python
a=-k*x/m
vx=vx-a*dt/2
for i in range(int(T/dt)):
    a0=a
    x+=vx*dt+1/2*a0*dt**2
    a=-k*x/m
    vx+=1/2*(a+a0)*dt
    t+=dt
    sol.loc[i]=[x,vx,t]
```

```python
plt.plot(sol['t'],sol['x'])
plt.plot(sol['t'], sol['x'][0]*np.sin(np.sqrt(k/m)*sol['t']))
```
